#üìå Extracci√≥n

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de los datos desde la API
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json'
df = pd.read_json(url)

# Normalizaci√≥n de datos
df = pd.json_normalize(df.to_dict(orient='records'))

# Se renombran las columnas para que coincidan con el diccionario
df = df.rename(columns={
    'customer.gender': 'gender',
    'customer.SeniorCitizen': 'SeniorCitizen',
    'customer.Partner': 'Partner',
    'customer.Dependents': 'Dependents',
    'customer.tenure': 'tenure',
    'phone.PhoneService': 'PhoneService',
    'phone.MultipleLines': 'MultipleLines',
    'internet.InternetService': 'InternetService',
    'internet.OnlineSecurity': 'OnlineSecurity',
    'internet.OnlineBackup': 'OnlineBackup',
    'internet.DeviceProtection': 'DeviceProtection',
    'internet.TechSupport': 'TechSupport',
    'internet.StreamingTV': 'StreamingTV',
    'internet.StreamingMovies': 'StreamingMovies',
    'account.Contract': 'Contract',
    'account.PaperlessBilling': 'PaperlessBilling',
    'account.PaymentMethod': 'PaymentMethod',
    'account.Charges.Monthly': 'MonthlyCharges',
    'account.Charges.Total': 'TotalCharges'
})

# Verificaci√≥n
print("Datos cargados y aplanados exitosamente. Forma del DataFrame:", df.shape)
print(df.columns)
df.head()
```

#üîß Transformaci√≥n

```python
# Exploraci√≥n del df.
print("Informaci√≥n del DataFrame:")
df.info()

print("\nTipos de datos por columna:")
print(df.dtypes)

# Identificaci√≥n de columnas relevantes para churn
relevant_columns = ['Churn', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Contract', 'PaymentMethod',
                    'InternetService', 'OnlineSecurity', 'TechSupport', 'SeniorCitizen']
print("\nColumnas relevantes para an√°lisis de evasi√≥n:", relevant_columns)
```

```python
# Convertir TotalCharges a num√©rico
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Valores ausentes
print("Valores ausentes por columna:")
print(df.isnull().sum())

# Duplicados
print("\nN√∫mero de duplicados:", df.duplicated().sum())

# Inconsistencias en categor√≠as
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nValores √∫nicos en {col}: {df[col].unique()}")
```

```python
# Manejo de ausentes en TotalCharges
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Eliminar duplicados
df = df.drop_duplicates()

# Estandarizar strings (lower case y strip para consistencia)
for col in categorical_cols:
    if col != 'customerID':
        df[col] = df[col].str.strip().str.lower()

# Manejo de "no internet service" / "no phone service" (mapear a 'no')
services = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for col in services:
    df[col] = df[col].replace({'no phone service': 'no', 'no internet service': 'no'})

# Verificaci√≥n
print("Datos despu√©s de limpieza:")
df.info()
print("\nValores ausentes restantes:", df.isnull().sum().sum())
```

```python
# Crear columna Cuentas_Diarias
df['Cuentas_Diarias'] = df['MonthlyCharges'] / 30
print("Nueva columna agregada:")
df[['MonthlyCharges', 'Cuentas_Diarias']].head()
```

```python
# Convertir Yes/No a 1/0
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0, ' ': 0})

# Renombrar columnas para claridad
df = df.rename(columns={'Charges.Monthly': 'MonthlyCharges', 'Charges.Total': 'TotalCharges'})

# Verificaci√≥n
df.head()
```

#üìä Carga y an√°lisis

```python
# An√°lisis descriptivo general
print("Descripci√≥n del DataFrame:")
print(df.describe(include='all'))

# Para num√©ricas espec√≠ficas
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
print("\nDescripci√≥n de columnas num√©ricas:")
print(df[numerical_cols].describe())

# Visualizaciones b√°sicas
plt.figure(figsize=(8, 4))
sns.histplot(df['tenure'], kde=True)
plt.title('Distribuci√≥n de Tenure')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df)
plt.title('Distribuci√≥n de Churn')
plt.show()
```

#üìÑInforme final

###Contexto del problema

Telecom X enfrenta un alto √≠ndice de evasi√≥n de clientes (churn), sin claridad previa sobre las causas que lo generan. El objetivo de este proyecto fue realizar un proceso de ETL (Extract, Transform, Load) y un an√°lisis exploratorio de datos para identificar patrones asociados a la evasi√≥n, entregando un dataset limpio y conclusiones relevantes al equipo de ciencia de datos.

###Tasa de churn

Considerando √∫nicamente los clientes con informaci√≥n de churn disponible, Telecom X presenta una tasa de evasi√≥n aproximada del 26,5%, lo que indica que m√°s de uno de cada cuatro clientes abandona la empresa. Este nivel de churn representa un riesgo significativo para la sostenibilidad del negocio.

üîç Principales hallazgos

###Tipo de contrato

El an√°lisis muestra que el tipo de contrato es el factor m√°s determinante en la evasi√≥n:

Contrato month-to-month: ‚âà 43% de churn

Contrato one year: ‚âà 11% de churn

Contrato two year: ‚âà 3% de churn

Los contratos mensuales presentan una tasa de evasi√≥n significativamente mayor, lo que sugiere que la falta de compromiso a largo plazo facilita la decisi√≥n de abandono.

###Antig√ºedad del cliente (tenure)

Los clientes que abandonan la empresa tienen, en promedio, una menor antig√ºedad, lo que indica que el churn ocurre principalmente en las primeras etapas de la relaci√≥n con el cliente.

###Cargos mensuales

Los clientes con churn presentan cargos mensuales ligeramente m√°s altos, lo que puede indicar una percepci√≥n de bajo valor frente al costo del servicio.

###Servicios adicionales

La ausencia de servicios adicionales como Tech Support se asocia a mayores tasas de churn, sugiriendo que el valor agregado contribuye a la retenci√≥n.

###Calidad de los datos

Se identificaron 224 registros sin informaci√≥n de churn, los cuales se mantuvieron como valores nulos para evitar introducir supuestos no fundamentados. Estos registros no fueron considerados en los an√°lisis comparativos.

###Conclusi√≥n general

El churn en Telecom X est√° fuertemente asociado a contratos flexibles, baja antig√ºedad y menor valor percibido del servicio. Estos factores deben ser priorizados en futuras estrategias de retenci√≥n y en modelos predictivos de evasi√≥n.

###Recomendaciones

Incentivar la migraci√≥n desde contratos mensuales a contratos de mayor duraci√≥n

Implementar estrategias de retenci√≥n temprana para clientes nuevos

Reforzar la oferta de servicios adicionales

Utilizar el dataset limpio para desarrollar modelos predictivos de churn
